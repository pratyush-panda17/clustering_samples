We will now calculate the mean Intra Group Similarity and mean inter group difference for each category. For every missing data point, it will be assigned the category it is closest to in terms of eucilidean distance. Thus each category is associated with a point (x,y) where x is the mean intra group similarity and y is the mean inter group difference. 

```python
predchar_means = {}
indexes = top_10_pred_char.index
avg_grp_diff = data.groupby('Predicted_Characteristics')['Inter_Group_Difference'].mean().to_dict()
avg_grp_sim = data.groupby('Predicted_Characteristics')['Intra_Group_Similarity'].mean().to_dict()
for index in indexes:
    predchar_means[index] = [avg_grp_sim[index],avg_grp_diff[index]]
```

```python
respro_means = {}
indexes = top_10_res_pro.index
avg_grp_diff = data.groupby('Resistance_Profile')['Inter_Group_Difference'].mean().to_dict()
avg_grp_sim = data.groupby('Resistance_Profile')['Intra_Group_Similarity'].mean().to_dict()
for index in indexes:
    respro_means[index] = [avg_grp_sim[index],avg_grp_diff[index]]
```

Code for eucilidean distance and classification. The classify function calculates the distance to each category and chooses the closest one.

```python
def distance(x,y):
    s = 0
    for i in range(len(x)):
        s+=math.pow(x[i]-y[i],2)
    return math.sqrt(s)
def classify(x,points):
    label = list(points.keys())[0]
    minimum = distance(x,points[label])
    for val in points:
        if  distance(x,points[val])<minimum:
            label = val
            minimum = distance(x,points[val])
    return label


```

Predicting all null values.

```python
for i in range(len(data)):
    if pd.isnull(data.iloc[i,-1]):
     data.iloc[i,-1] = classify([data.iloc[i,4],data.iloc[i,5]],predchar_means)
        
    if pd.isnull(data.iloc[i,-2]):
        data.iloc[i,-2] = classify([data.iloc[i,4],data.iloc[i,5]],respro_means) 
```

The entry date column is another categorical variable with many categories. But the entry date does give us crucial infomration as similar samples would most likely be taken on the same day. To calculate the distance between categorical points we will add 1 if the categories are unequal, 0 otherwise. It is a special kind of eucilidean distance function.Thus to avoid biases, we need to normalize our data, as the distance between every categorical variable is 1 or 0.

```python
categories = []
for i in range(5):
    categories.append(data[column_names[i]].unique())
```

The clustering class implements the K means clustering algorithm. It uses the same distance function that we defined earlier.

```python
class Kmeansclustering:
    def __init__(self,k=3,max_iters = 100):
        self.k = k
        self.centroids = None
        self.max_iters = max_iters

    def random_point():
        return [category[random.randint(0,len(category)-1)] for category in categories] + [random.random() for x in range(3)]

    def distance_fun(x,y):
        assert len(x)==len(y),"Dimensions of both points are unequal, cannot calculate distance"
        s=0
        for i in range(len(x)):
            if issubclass(type(x[i]),str):
                s+=0 if x[i]==y[i] else 1
                continue
            s+=math.pow(x[i]-y[i],2)
        return math.sqrt(s)
        
    def intraclusterdis(cluster):
        N = len(cluster)
        k = N//22
        s= 0
        for i in range(k):
            for j in range(i+1,k):
                point1 = cluster.iloc[i,:].values.flatten().tolist()
                point2 = cluster.iloc[j,:].values.flatten().tolist()
                s+= Kmeansclustering.distance_fun(point1,point2)
        return s/math.comb(k,2)

    def sillouette_score(clusters,centroids):
        N = len(clusters)
        s=0
        for cluster in clusters:
            s+=Kmeansclustering.intraclusterdis(cluster)
            print("finished cluster")
        a = s/N
    
        s=0
        for i in range(N):
            for j in range(i+1,N):
                s+=Kmeansclustering.distance_fun(centroids[i],centroids[j])
    
        b = s/math.comb(N,2)
        return (b-a)/(max(a,b))
        
    def norm(X):
        s=0
        for i in X:
            s+=i**2
        return math.sqrt(s)
        
    def train(self,df):
        N = len(df)
        self.centroids = [Kmeansclustering.random_point() for x in range(self.k)] #generating random centroids

        for i in range(self.max_iters):
            print(i)
            labels = []
            for i in range(N): #classifying each point based on the closest centroid
                row = df.loc[i, :].values.flatten().tolist()
                distances = np.array([Kmeansclustering.distance_fun(row,centroid) for centroid in self.centroids])
                labels.append(np.argmin(distances))
            labels = np.array(labels) #contains the cluster of the ith row at the ith index

            cluster_indices = []
            new_centroids = []
        
            for i in range(self.k): 
                cluster_indices.append(np.argwhere(labels==i).flatten()) #the ith index consists of all the rows which are classifed as cluster i

            for i,indices in enumerate(cluster_indices):
                if len(indices)==0:
                    new_centroids.append(self.centroids[i])
                else:
                    cluster_df = df.iloc[indices]
                    cluster_df
                    new_centroids.append([cluster_df.iloc[:,j].value_counts().iloc[0] for j in range(5)] + [cluster_df.iloc[:,j].mean() for j in range(5,8)])
            #calculating the new centroid as the mode of categorical data and mean of numerical data
            movement =[Kmeansclustering.distance_fun(centroids[i],new_centroids[i]) for i in range(self.k)]
            if Kmeansclustering.norm(movement) <=0.00001:
                break 
            self.centroids = new_centroids

        return cluster_indices,self.centroids
    
```

Training a model through 100 iterations.
